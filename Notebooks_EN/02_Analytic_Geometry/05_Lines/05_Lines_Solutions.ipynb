# Solutions to Tasks Related to Equations of Lines on a Plane

### 1. The line passes through points \( A(1, 2) \) and \( B(3, 4) \). Find the equation of the line.
The slope of the line:
$$
m = \frac{y_2 - y_1}{x_2 - x_1} = \frac{4 - 2}{3 - 1} = 1
$$
Using the point-slope form:
$$
y - 2 = 1(x - 1) \implies y = x + 1
$$

### 2. The line passes through point \( A(1, 2) \) and is parallel to the line \( y = 2x + 3 \). Find the equation of the line.
The slope of the line is the same, \( m = 2 \). Using the point-slope form:
$$
y - 2 = 2(x - 1) \implies y = 2x
$$

### 3. The line passes through point \( A(1, 2) \) and is perpendicular to the line \( y = 2x + 3 \). Find the equation of the line.
The slope of a line perpendicular to \( m = 2 \) is $ m = -\frac{1}{2} $ Using the point-slope form:
$$
y - 2 = -\frac{1}{2}(x - 1) \implies y = -\frac{1}{2}x + \frac{5}{2}
$$

### 4. Two lines \( y = 2x + 3 \) and \( y = 3x + 2 \). Find their intersection point and the angle between them.
#### Intersection:
Solve:
$$
2x + 3 = 3x + 2 \implies x = 1, \, y = 5
$$
Intersection point: \( (1, 5) \).

#### Angle:
The angle between the lines is given by:
$$
\tan\theta = \left|\frac{m_1 - m_2}{1 + m_1m_2}\right| = \left|\frac{2 - 3}{1 + 2 \cdot 3}\right| = \frac{1}{7}
$$
$theta = \arctan\left(\frac{1}{7}\right)$

### 5. Write the equation of the line passing through \( A(1, 2) \) and parallel to the vector $ \vec{v} = [2, 3] $.
The direction vector gives the slope $ m = \frac{3}{2} $. Using the point-slope form:
$$
y - 2 = \frac{3}{2}(x - 1) \implies y = \frac{3}{2}x + \frac{1}{2}
$$

### 6. Line \( y = 2x + 3 \). Find a perpendicular and a parallel line.
#### Parallel:
$$
y = 2x + c, \quad \text{for any } c \neq 3.
$$

#### Perpendicular:
$$
y = -\frac{1}{2}x + c, \quad \text{for any } c.
$$

### 7. Line \( y = 2x + 3 \) and point \( A(1, 2) \). Find the distance from \( A \) to the line.
The distance formula:
$$
d = \frac{|2(1) - 1(2) + 3|}{\sqrt{2^2 + 1^2}} = \frac{|2 + 3|}{\sqrt{5}} = \frac{5}{\sqrt{5}} = \sqrt{5}
$$

### 8. Line intersects axes at \( A(2, 0) \) and \( B(0, 3) \). Find the equation of the line.
Slope:
$$
m = \frac{3 - 0}{0 - 2} = -\frac{3}{2}
$$
Equation:
$$
y - 0 = -\frac{3}{2}(x - 2) \implies y = -\frac{3}{2}x + 3
$$

### 9. Calculate the angle between \( y = x + 3 \) and the \( 0x \)-axis.
The slope \( m = 1 \). Angle:
$$
\theta = \arctan(1) = 45^\circ
$$

### 10. Provide a vector perpendicular to the line \( x + y + 1 = 0 \).
The normal vector of the line is:
$$
\vec{n} = [1, 1].
