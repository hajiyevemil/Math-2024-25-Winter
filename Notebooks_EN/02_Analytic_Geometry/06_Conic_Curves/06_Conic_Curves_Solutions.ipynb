# Solutions to Tasks Related to Conic Sections

### 1. Find the equation of a circle with center at point \( A(1, 2) \) and radius \( r = 3 \).
The general equation of a circle is:
$$
(x - h)^2 + (y - k)^2 = r^2
$$
Here, \( h = 1 \), \( k = 2 \), and \( r = 3 \):
$$
(x - 1)^2 + (y - 2)^2 = 9
$$

---

### 2. Find the equation of a parabola intersecting the \( 0x \)-axis at points \( x = 2, x = 4 \), and passing through point \( y(3) = 1 \).
The general form of a parabola is:
$$
y = a(x - x_1)(x - x_2)
$$
Substitute \( x_1 = 2 \), \( x_2 = 4 \):
$$
y = a(x - 2)(x - 4)
$$
Using \( y(3) = 1 \):
$$
1 = a(3 - 2)(3 - 4) \implies 1 = a(-1) \implies a = -1
$$
The equation of the parabola is:
$$
y = -(x - 2)(x - 4) \implies y = -x^2 + 6x - 8
$$

---

### 3. Find the center of the ellipse with the equation \( x^2 + 4y^2 - 4x - 16y + 16 = 0 \).
Rewrite the equation in standard form by completing the square:
1. Group \( x \)-terms and \( y \)-terms:
   $$
   (x^2 - 4x) + (4y^2 - 16y) = -16
   $$
2. Complete the square for \( x^2 - 4x \):
   $$
   x^2 - 4x = (x - 2)^2 - 4
   $$
3. Complete the square for \( 4y^2 - 16y \):
   $$
   4(y^2 - 4y) = 4((y - 2)^2 - 4) = 4(y - 2)^2 - 16
   $$
4. Substitute back:
   $$
   (x - 2)^2 - 4 + 4(y - 2)^2 - 16 = -16 \implies (x - 2)^2 + 4(y - 2)^2 = 36
   $$
The center of the ellipse is:
$$
(2, 2)
$$

---

### 4. Find the slope (\( m > 0 \)) of the line \( y = mx - 5 \) that is tangent to the circle with the equation \( x^2 + y^2 = 1 \).
The distance from the center of the circle \((0, 0)\) to the line must equal the radius:
$$
\text{Distance} = \frac{|c|}{\sqrt{1 + m^2}} = 1, \quad \text{where } c = -5.
$$
Solve:
$$
\frac{|-5|}{\sqrt{1 + m^2}} = 1 \implies \sqrt{1 + m^2} = 5 \implies 1 + m^2 = 25 \implies m^2 = 24 \implies m = \sqrt{24} = 2\sqrt{6}.
$$

---

### 5. Find the intersection points of the hyperbola \( x^2 - y^2 = 1 \) with the ellipse \( x^2 + 4y^2 = 6 \).
Solve the system of equations:
1. From \( x^2 - y^2 = 1 \), express \( x^2 \):
   $$
   x^2 = y^2 + 1
   $$
2. Substitute into \( x^2 + 4y^2 = 6 \):
   $$
   (y^2 + 1) + 4y^2 = 6 \implies 5y^2 + 1 = 6 \implies 5y^2 = 5 \implies y^2 = 1 \implies y = \pm 1
   $$
3. Substitute \( y = 1 \) and \( y = -1 \) into \( x^2 = y^2 + 1 \):
   $$
   x^2 = 1 + 1 = 2 \implies x = \pm\sqrt{2}.
   $$
The intersection points are:
$$
(\sqrt{2}, 1), (-\sqrt{2}, 1), (\sqrt{2}, -1), (-\sqrt{2}, -1).
$$

---

### 6. For the given hyperbola \( x^2 - y^2 = 1 \), find the distance between its branches.
The distance between the branches of a hyperbola along the transverse axis is:
$$
2a, \quad \text{where } a = \sqrt{1}.
$$
Here, \( a = 1 \), so the distance is:
$$
2a = 2.
$$
