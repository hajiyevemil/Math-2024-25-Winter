# Solutions to Tasks Related to Equations of Planes in Space

### 1. The plane passes through points $A(1, 2, 3)$, $B(3, 4, 5)$, and $C(2, 1, 4)$. Find the equation of the plane.
The equation of the plane is:
$$
\vec{n} \cdot (\vec{r} - \vec{r}_0) = 0
$$
where $\vec{n} = \vec{AB} \times \vec{AC}$.

1. Calculate $\vec{AB} = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2)$.
2. Calculate $\vec{AC} = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1)$.
3. Compute the cross product:
   $$
   \vec{AB} \times \vec{AC} =
   \begin{vmatrix}
   \mathbf{i} & \mathbf{j} & \mathbf{k} \\
   2 & 2 & 2 \\
   1 & -1 & 1
   \end{vmatrix}
   = (4 - (-2))\mathbf{i} - (2 - 2)\mathbf{j} + (2 - (-2))\mathbf{k} = (6, 0, 4).
   $$
4. The equation becomes:
   $$
   6(x - 1) + 0(y - 2) + 4(z - 3) = 0 \quad \Rightarrow \quad 6x + 4z = 18.
   $$

### 2. The plane passes through point $A(1, 2, 3)$ and is parallel to the plane $2x + 3y + 4z = 5$. Find the equation of the plane.
If the plane is parallel, the normal vector is the same: $\vec{n} = (2, 3, 4)$. Use the point-normal form:
$$
2(x - 1) + 3(y - 2) + 4(z - 3) = 0 \quad \Rightarrow \quad 2x + 3y + 4z = 20.
$$

### 3. The plane passes through point $A(1, 2, 3)$ and is perpendicular to the normal vector $\vec{n} = [2, 3, 4]$. Find the equation of the plane.
The normal vector directly gives the equation:
$$
2(x - 1) + 3(y - 2) + 4(z - 3) = 0 \quad \Rightarrow \quad 2x + 3y + 4z = 20.
$$

### 4. Find the line of intersection of the planes $2x + 3y + 4z = 5$ and $3x + 4y + 2z = 6$.
1. Solve the system:
   $$
   \begin{cases}
   2x + 3y + 4z = 5 \\
   3x + 4y + 2z = 6
   \end{cases}.
   $$
2. Parameterize by eliminating one variable (e.g., $z = t$):
   - Solve for $x$ and $y$ in terms of $t$.
   - The line is of the form:
     $$
     \vec{r}(t) = \vec{r}_0 + t\vec{d}.
     $$

### 5. Write the equation of the plane passing through point $A(1, 2, 3)$ and parallel to vectors $\vec{v}_1 = [1, 0, 1]$ and $\vec{v}_2 = [0, 1, -1]$.
The normal vector is $\vec{v}_1 \times \vec{v}_2$:
$$
\vec{n} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 0 & 1 \\
0 & 1 & -1
\end{vmatrix}
= (-1, 1, 1).
$$
The equation becomes:
$$
-1(x - 1) + 1(y - 2) + 1(z - 3) = 0 \quad \Rightarrow \quad -x + y + z = 4.
$$

### 6. Find an example of a plane parallel and perpendicular to $2x + 3y + 4z = 5$.
- **Parallel**: Keep the same normal vector, change the constant:
  $$
  2x + 3y + 4z = 10.
  $$
- **Perpendicular**: Choose a plane with a normal vector orthogonal to $(2, 3, 4)$, such as $(4, -2, 0)$. An example equation is:
  $$
  4x - 2y = 7.
  $$

### 7. Find the distance from point $A(1, 2, 3)$ to the plane $2x + 3y + 4z = 5$.
The distance is given by:
$$
d = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}} = \frac{|2 + 6 + 12 - 5|}{\sqrt{4 + 9 + 16}} = \frac{15}{\sqrt{29}}.
$$

### 8. The plane intersects the coordinate axes at points $A(2, 0, 0)$, $B(0, 3, 0)$, and $C(0, 0, 4)$. Find the equation of the plane.
The intercept form is:
$$
\frac{x}{2} + \frac{y}{3} + \frac{z}{4} = 1 \quad \Rightarrow \quad 6x + 4y + 3z = 12.
$$

### 9. Calculate the angle between the plane $x + y + z = 1$ and the plane $x = 0$ (i.e., the $yz$-plane).
The angle is given by:
$$
\cos\theta = \frac{\vec{n}_1 \cdot \vec{n}_2}{\|\vec{n}_1\| \|\vec{n}_2\|},
$$
where $\vec{n}_1 = (1, 1, 1)$ and $\vec{n}_2 = (1, 0, 0)$.
$$
\cos\theta = \frac{1(1) + 1(0) + 1(0)}{\sqrt{1^2 + 1^2 + 1^2} \sqrt{1^2}} = \frac{1}{\sqrt{3}} \quad \Rightarrow \quad \theta = \cos^{-1}\left(\frac{1}{\sqrt{3}}\right).
$$

### 10. Find the vector perpendicular to the plane $x + y + z = 1$.
The vector perpendicular to the plane is the normal vector:
$$
\vec{n} = (1, 1, 1).
$$
