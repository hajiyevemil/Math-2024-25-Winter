# Solutions to Tasks Related to Second-Order Surfaces

### 1. Write the equation of a sphere with center at point $P = (1, 2, 3)$ and radius $r = 3$.
The equation of the sphere is:
$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9.
$$

### 2. Do the spheres with equations $x^2 + y^2 + z^2 = 1$ and $x^2 + y^2 + z^2 = 2$ have any common points?
1. The two spheres are concentric with center at $(0, 0, 0)$ and have radii $r_1 = 1$ and $r_2 = \sqrt{2}$.
2. Since the radii are different, and they share the same center, the spheres do not intersect.
   
**Conclusion:** The spheres do not have any common points.

---

### 3. What curve in space is formed by the intersection of the sphere $x^2 + y^2 + z^2 = 1$ with the sphere $(x - 1)^2 + y^2 + z^2 = 1$? Find the equation of this curve.
1. Expand the second sphere:
   $$
   (x - 1)^2 + y^2 + z^2 = 1 \quad \Rightarrow \quad x^2 - 2x + 1 + y^2 + z^2 = 1.
   $$
   Simplify:
   $$
   x^2 + y^2 + z^2 - 2x = 0.
   $$
2. Substitute $x^2 + y^2 + z^2 = 1$ into $x^2 + y^2 + z^2 - 2x = 0$:
   $$
   1 - 2x = 0 \quad \Rightarrow \quad x = \frac{1}{2}.
   $$
3. The curve lies in the plane $x = \frac{1}{2}$ and on the sphere $x^2 + y^2 + z^2 = 1$. The equation of the curve is:
   $$
   \left(\frac{1}{2}\right)^2 + y^2 + z^2 = 1 \quad \Rightarrow \quad y^2 + z^2 = \frac{3}{4}.
   $$
   This is a circle with center $\left(\frac{1}{2}, 0, 0\right)$ and radius $\frac{\sqrt{3}}{2}$.

### 4. Write the equation of the tangent plane to the paraboloid $z = (x - 1)^2 + y^2 + 1$ at point $P = (1, 0, 1)$.
1. The equation of the paraboloid is:
   $$
   z = (x - 1)^2 + y^2 + 1.
   $$
2. Compute partial derivatives:
   $$
   \frac{\partial z}{\partial x} = 2(x - 1), \quad \frac{\partial z}{\partial y} = 2y.
   $$
3. At $P = (1, 0, 1)$:
   $$
   \frac{\partial z}{\partial x} = 2(1 - 1) = 0, \quad \frac{\partial z}{\partial y} = 2(0) = 0.
   $$
4. The tangent plane is:
   $$
   z - z_0 = \frac{\partial z}{\partial x}(x - x_0) + \frac{\partial z}{\partial y}(y - y_0),
   $$
   where $(x_0, y_0, z_0) = (1, 0, 1)$. Substituting:
   $$
   z - 1 = 0(x - 1) + 0(y - 0) \quad \Rightarrow \quad z = 1.
   $$

**Conclusion:** The tangent plane is $z = 1$.
