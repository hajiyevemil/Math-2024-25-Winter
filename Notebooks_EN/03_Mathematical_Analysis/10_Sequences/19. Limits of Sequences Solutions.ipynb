# Limits of Sequences

### Problem 1
**Calculate:**

1. $\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n}$:  
   $\frac{n^2 + 3n}{2n^2 - 2n} = \frac{n^2(1 + \frac{3}{n})}{n^2(2 - \frac{2}{n})} = \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}$  
   As $n \to \infty$, $\frac{3}{n} \to 0$ and $\frac{2}{n} \to 0$, so:  
   $\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n} = \frac{1}{2}$  

2. $\lim_{n \to \infty} \frac{(2n + 3)^3}{n^3 - 1}$:  
   Expand $(2n + 3)^3$:  
   $(2n + 3)^3 = 8n^3 + 12n^2 \cdot 3 + 6n \cdot 9 + 27 = 8n^3 + 36n^2 + 54n + 27$  
   The fraction becomes:  
   $\frac{8n^3 + 36n^2 + 54n + 27}{n^3 - 1}$  
   Divide numerator and denominator by $n^3$:  
   $\frac{8 + \frac{36}{n} + \frac{54}{n^2} + \frac{27}{n^3}}{1 - \frac{1}{n^3}}$  
   As $n \to \infty$, all terms involving $\frac{1}{n^k}$ tend to $0$, so:  
   $\lim_{n \to \infty} \frac{(2n + 3)^3}{n^3 - 1} = 8$

### Problem 2
**Prove using the squeeze theorem:**  

$\lim_{n \to \infty} \frac{\sin(n)}{n}$:  

- $\sin(n)$ is bounded: $-1 \leq \sin(n) \leq 1$  
- Divide through by $n$ (for $n > 0$):  
  $-\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}$  
- As $n \to \infty$, both $-\frac{1}{n}$ and $\frac{1}{n}$ approach $0$, so by the squeeze theorem:  
  $\lim_{n \to \infty} \frac{\sin(n)}{n} = 0$

### Problem 3
**Find the limit of the sequence:**  

$a_n = \left(1 + \frac{1}{n}\right)^n$:  

- Recognize this as the definition of $e$:  
  $\lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e$  

So:  
$\lim_{n \to \infty} a_n = e$
