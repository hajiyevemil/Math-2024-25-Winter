### Problem 1
**Draw in a single GeoGebra notebook the following functions:**

- $f(x) = x^2$
- $g(x) = \sqrt{x}$
- $h(x) = \frac{1}{x}$
- $j(x) = \sin(x)$

**Find the value of all the above functions at $x = 2$:**

- $f(2) = 2^2 = 4$
- $g(2) = \sqrt{2}$
- $h(2) = \frac{1}{2} = 0.5$
- $j(2) = \sin(2)$

### Problem 2
Let $f(x) = 3x - 1$ and $g(x) = \sqrt{x}$. Find:

1. $f(g(x))$:
   \[
   f(g(x)) = f(\sqrt{x}) = 3\sqrt{x} - 1
   \]
2. $g(f(x))$:
   \[
   g(f(x)) = g(3x - 1) = \sqrt{3x - 1}
   \]
3. $f(f(x))$:
   \[
   f(f(x)) = f(3x - 1) = 3(3x - 1) - 1 = 9x - 3 - 1 = 9x - 4
   \]
4. $g(g(x))$:
   \[
   g(g(x)) = g(\sqrt{x}) = \sqrt{\sqrt{x}} = x^{1/4}
   \]

**Visualize functions in a single GeoGebra notebook.**

### Problem 3
Let $f(x) = e^x$ and $g(x) = \ln(x)$. Check $f(g(x))$ and $g(f(x))$:

- $f(g(x)) = e^{\ln(x)} = x$
- $g(f(x)) = \ln(e^x) = x$

**Observation:** $f(g(x))$ and $g(f(x))$ both result in $x$, confirming that $f(x)$ and $g(x)$ are inverses.

### Problem 5

- The function $f$ is **not invertible** because the $y$-value $7$ is repeated for different $x$-values. A function must be one-to-one to have an inverse.

### Problem 6
We have the function $f(x) = x - 1$. Find the inverse function $f^{-1}$:

1. Start with $y = x - 1$.
2. Solve for $x$: $x = y + 1$.
3. Swap $x$ and $y$: $f^{-1}(x) = x + 1$.

**Visualization:** Plot both $f(x)$ and $f^{-1}(x)$ in the same GeoGebra notebook to observe symmetry along the line $y = x$.
