# Limits of Real Functions

### Problem 1
**Compute:**  

1. $\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}$:  
   Divide numerator and denominator by $x^3$:  
   $\frac{x^3 + 2x^2}{x^4 - 3x^3} = \frac{1 + \frac{2}{x}}{x - 3}$  
   As $x \to \infty$, $\frac{2}{x} \to 0$:  
   $\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} = \frac{1}{\infty - 3} = 0$  

---

### Problem 2
**Find:**  

$\lim_{x \to 0} \frac{\sin(3x)}{2x + 1}$:  
- Numerator: $\sin(3x) \approx 3x$ as $x \to 0$  
- Denominator: $2x + 1 \approx 1$ as $x \to 0$  
Thus:  
$\lim_{x \to 0} \frac{\sin(3x)}{2x + 1} = \frac{3x}{1} = 0$

### Problem 3
**Find the asymptotes of the functions:**  

1. $f(x) = \frac{x^2 - 1}{x^2 + 1}$:  
   - Horizontal asymptote:  
     As $x \to \infty$, divide numerator and denominator by $x^2$:  
     $f(x) = \frac{1 - \frac{1}{x^2}}{1 + \frac{1}{x^2}} \to \frac{1}{1} = 1$  
     So, the horizontal asymptote is $y = 1$.  

   - Vertical asymptote:  
     $f(x)$ has no vertical asymptotes since $x^2 + 1 \neq 0$ for all $x$.  

2. $g(x) = \frac{\sin(x)}{x^2 + 1}$:  
   - Horizontal asymptote:  
     As $x \to \infty$, $\sin(x)$ oscillates between $-1$ and $1$, and $x^2 + 1 \to \infty$:  
     $\frac{\sin(x)}{x^2 + 1} \to 0$  
     So, the horizontal asymptote is $y = 0$.  

   - Vertical asymptote:  
     $g(x)$ has no vertical asymptotes since $x^2 + 1 \neq 0$ for all $x$.
