## Solutions to Derivatives Problems

### 1. Compute derivatives of functions:

1. $$ y(x) = -3x + 3 \quad \Rightarrow \quad y'(x) = -3 $$
2. $$ y(x) = \pi x + \sin(1) \quad \Rightarrow \quad y'(x) = \pi $$
3. $$ y(x) = 4 + \sin(2) \quad \Rightarrow \quad y'(x) = 0 $$
4. $$ y(x) = 2x^3 - 3x^2 + 8x - 9 \quad \Rightarrow \quad y'(x) = 6x^2 - 6x + 8 $$
5. $$ y(x) = 6x^{1/3} \quad \Rightarrow \quad y'(x) = 2x^{-2/3} $$
6. $$ y(x) = \sqrt{x} \quad \Rightarrow \quad y'(x) = \frac{1}{2\sqrt{x}} $$
7. $$ y(x) = \cos(x) + \sin(x) \quad \Rightarrow \quad y'(x) = -\sin(x) + \cos(x) $$
8. $$ y(x) = 2\sin(x)\cos(x) \quad \Rightarrow \quad y'(x) = 2\cos^2(x) - 2\sin^2(x) $$
9. $$ y(x) = x\sin(x) \quad \Rightarrow \quad y'(x) = \sin(x) + x\cos(x) $$
10. $$ y(x) = (x + 1)(x + 1) \quad \Rightarrow \quad y'(x) = 2(x + 1) $$
11. $$ y(x) = \frac{x}{x + 1} \quad \Rightarrow \quad y'(x) = \frac{1}{(x + 1)^2} $$
12. $$ y(x) = (x + 1)\exp(x) \quad \Rightarrow \quad y'(x) = (x + 1)\exp(x) + \exp(x) $$
13. $$ y(x) = \sin(x^2) \quad \Rightarrow \quad y'(x) = 2x\cos(x^2) $$
14. $$ y(x) = \exp(-2x) \quad \Rightarrow \quad y'(x) = -2\exp(-2x) $$
15. $$ y(x) = \frac{1}{\sin(x) + 1} \quad \Rightarrow \quad y'(x) = -\frac{\cos(x)}{(\sin(x) + 1)^2} $$
16. $$ y(x) = \sqrt{2x + 1} \quad \Rightarrow \quad y'(x) = \frac{1}{\sqrt{2x + 1}} $$

### 2. Prove:
**Prove**: $$ \frac{d}{dx}(\ln(\sin(x))) = \cot(x) $$

**Solution**:  
Using the chain rule:
$$ \frac{d}{dx}(\ln(\sin(x))) = \frac{1}{\sin(x)} \cdot \cos(x) = \cot(x) $$

### 3. For $$ f(x) = \cos(x) $$, verify that $$ f'(x) = -f(x) $$.

**Solution**:  
The derivative of $$ f(x) = \cos(x) $$ is:
$$ f'(x) = -\sin(x) $$.  
Since $$ f'(x) = -\cos(x) $$, this verifies that $$ f'(x) = -f(x) $$.

### 4. Using de l'Hospital's Rule, find the improper limits:

1. $$ \lim_{x \to 0} \frac{\sin(x)}{x} \quad \Rightarrow \quad \text{Apply L'Hospital's Rule: } \frac{\cos(x)}{1} = 1. $$  
   So, $$ \lim_{x \to 0} \frac{\sin(x)}{x} = 1 $$.

2. $$ \lim_{x \to \infty} \frac{\ln(x)}{x} \quad \Rightarrow \quad \text{Apply L'Hospital's Rule: } \frac{1/x}{1} = \frac{1}{x} \to 0. $$  
   So, $$ \lim_{x \to \infty} \frac{\ln(x)}{x} = 0 $$.

3. $$ \lim_{x \to \infty} \frac{\exp(x)}{x} \quad \Rightarrow \quad \text{Growth of } \exp(x) \text{ dominates, so } \infty. $$  
   So, $$ \lim_{x \to \infty} \frac{\exp(x)}{x} = \infty $$.

### 5. In physics, the position of a particle is given by $$ x(t) = 3t^2 - 6t + 1 $$.  
Find the velocity $$ V(t) = x'(t) $$ and acceleration $$ a(t) = V'(t) $$ of the particle at time $$ t = 2 $$.

**Solution**:
1. Velocity:
   $$ V(t) = \frac{dx}{dt} = 6t - 6. $$  
   At $$ t = 2 $$:  
   $$ V(2) = 6(2) - 6 = 12 - 6 = 6. $$

2. Acceleration:
   $$ a(t) = \frac{dV}{dt} = 6. $$  
   At $$ t = 2 $$:  
   $$ a(2) = 6. $$
