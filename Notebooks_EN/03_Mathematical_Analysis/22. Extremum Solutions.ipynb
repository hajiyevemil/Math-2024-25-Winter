## Solutions to Extremum Problems

### 6. The profit function is $$ P(u) = -2u^2 + 50u - 300 $$, where $$ u $$ is the number of units sold.  
Find the number of units that maximize profit.

**Solution**:  
To maximize the profit, find the critical points by taking the derivative and setting it to zero:  
$$ P'(u) = \frac{d}{du}(-2u^2 + 50u - 300) = -4u + 50. $$  
Set $$ P'(u) = 0 $$:  
$$ -4u + 50 = 0 \quad \Rightarrow \quad u = \frac{50}{4} = 12.5. $$  

To confirm that this is a maximum, evaluate the second derivative:  
$$ P''(u) = -4. $$  
Since $$ P''(u) < 0 $$, the function has a maximum at $$ u = 12.5. $$  

Thus, the number of units that maximizes profit is:  
$$ u = 12.5. $$

### 7. You have 10 meters of string, and you need to use it to enclose the largest possible rectangle.  
Find the dimensions of the rectangle.

**Solution**:  
Let the length of the rectangle be $$ l $$ and the width be $$ w $$. The perimeter is:  
$$ 2l + 2w = 10 \quad \Rightarrow \quad l + w = 5. $$  
The area is:  
$$ A = l \cdot w. $$  

Substitute $$ w = 5 - l $$ into the area formula:  
$$ A = l(5 - l) = 5l - l^2. $$  

To maximize the area, take the derivative of $$ A $$ and set it to zero:  
$$ \frac{dA}{dl} = 5 - 2l. $$  
Set $$ \frac{dA}{dl} = 0 $$:  
$$ 5 - 2l = 0 \quad \Rightarrow \quad l = \frac{5}{2} = 2.5. $$  

Thus, $$ w = 5 - l = 5 - 2.5 = 2.5. $$  

The dimensions of the rectangle that maximize the area are:  
$$ l = 2.5 \, \text{m}, \, w = 2.5 \, \text{m}. $$

### 8. Find the extremum of $$ f(x) = x^2 + 3x - 5 $$.

**Solution**:  
Take the derivative of $$ f(x) $$:  
$$ f'(x) = 2x + 3. $$  

Set $$ f'(x) = 0 $$ to find critical points:  
$$ 2x + 3 = 0 \quad \Rightarrow \quad x = -\frac{3}{2}. $$  

To classify the critical point, take the second derivative:  
$$ f''(x) = 2. $$  
Since $$ f''(x) > 0 $$, the function has a minimum at $$ x = -\frac{3}{2}. $$  

Substitute $$ x = -\frac{3}{2} $$ into the original function:  
$$ f\left(-\frac{3}{2}\right) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5 = \frac{9}{4} - \frac{9}{2} - 5 = -\frac{29}{4}. $$  

The minimum value is:  
$$ f(x) = -\frac{29}{4} \, \text{at} \, x = -\frac{3}{2}. $$

### 9. Find the extremum of $$ f(x) = \frac{x^2 + 2x + 1}{x - 1} $$.

**Solution**:  
Let $$ f(x) = \frac{x^2 + 2x + 1}{x - 1} = \frac{(x + 1)^2}{x - 1}. $$  

Take the derivative using the quotient rule:  
$$ f'(x) = \frac{(x - 1) \cdot 2(x + 1) - (x + 1)^2 \cdot 1}{(x - 1)^2}. $$  

Simplify:  
$$ f'(x) = \frac{2(x + 1)(x - 1) - (x + 1)^2}{(x - 1)^2} = \frac{(x + 1)\left[2(x - 1) - (x + 1)\right]}{(x - 1)^2}. $$  
$$ f'(x) = \frac{(x + 1)(x - 3)}{(x - 1)^2}. $$  

Set $$ f'(x) = 0 $$:  
$$ (x + 1)(x - 3) = 0 \quad \Rightarrow \quad x = -1 \, \text{or} \, x = 3. $$  

Check the second derivative or test intervals to classify these points.  
At $$ x = -1 $$, the function is undefined due to the denominator.  
At $$ x = 3 $$, substitute into the original function:  
$$ f(3) = \frac{3^2 + 2(3) + 1}{3 - 1} = \frac{9 + 6 + 1}{2} = \frac{16}{2} = 8. $$  

The extremum is:  
Maximum value: $$ f(x) = 8 \, \text{at} \, x = 3. $$
