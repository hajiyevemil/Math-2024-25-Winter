## Solutions to Taylor Series Problems

### 1. Find the Taylor series and visualize the obtained functions in GeoGebra:

#### (a) $$ f(x) = \cos(x) $$ around $$ x = 0 $$ up to the 4th degree.

The Taylor series expansion of $$ \cos(x) $$ around $$ x = 0 $$ is:  
$$ \cos(x) = 1 - \frac{x^2}{2!} + \frac{x^4}{4!} + \mathcal{O}(x^6). $$  

Up to the 4th degree:  
$$ \cos(x) \approx 1 - \frac{x^2}{2} + \frac{x^4}{24}. $$

---

#### (b) $$ h(x) = \frac{1}{1 - x} $$ around $$ x = 0 $$ up to the 4th degree.

The Taylor series expansion of $$ \frac{1}{1 - x} $$ around $$ x = 0 $$ is:  
$$ \frac{1}{1 - x} = 1 + x + x^2 + x^3 + x^4 + \mathcal{O}(x^5). $$  

Up to the 4th degree:  
$$ \frac{1}{1 - x} \approx 1 + x + x^2 + x^3 + x^4. $$

---

#### (c) $$ g(x) = \sin(x) $$ around $$ x = \pi $$ up to the 4th degree.

To expand $$ \sin(x) $$ around $$ x = \pi $$, substitute $$ u = x - \pi $$ so that $$ \sin(x) = \sin(\pi + u). $$  
Using the trigonometric identity $$ \sin(\pi + u) = -\sin(u) $$, the expansion is:  
$$ \sin(x) = -u + \frac{u^3}{3!} + \mathcal{O}(u^5), \quad \text{where} \, u = x - \pi. $$  

Up to the 4th degree:  
$$ \sin(x) \approx -(x - \pi) + \frac{(x - \pi)^3}{6}. $$

---

### 2. Find a tangent line $$ y = f'(x_0)(x - x_0) + f(x_0) $$ to the function $$ f(x) = e^{\sin(x)} $$ at $$ x_0 = \pi $$.

**Solution**:  
1. Compute $$ f(x) = e^{\sin(x)} $$ at $$ x_0 = \pi $$:  
   $$ f(\pi) = e^{\sin(\pi)} = e^0 = 1. $$  

2. Compute the derivative $$ f'(x) $$:  
   $$ f'(x) = e^{\sin(x)} \cdot \cos(x). $$  

   At $$ x_0 = \pi $$:  
   $$ f'(\pi) = e^{\sin(\pi)} \cdot \cos(\pi) = 1 \cdot (-1) = -1. $$  

3. The equation of the tangent line is:  
   $$ y = f'(\pi)(x - \pi) + f(\pi). $$  
   Substituting values:  
   $$ y = -1(x - \pi) + 1. $$  

   Simplify:  
   $$ y = -x + \pi + 1. $$

The tangent line is:  
$$ y = -x + \pi + 1. $$
